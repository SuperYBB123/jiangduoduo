```python
train=dict()
```

```python
data0=pd.read_csv(r'C:\Users\JDD\query-impala-310822.csv',encoding="gbk")
data0
```

```python
data0.drop(['essay_id.1','c_title','d_dateline'],axis=1,inplace=True)
data0.to_csv(r'C:\Users\JDD\query-impala-310822_7.csv',index=False)
```

```python
for line in open(r'C:\Users\JDD\query-impala-310822_7.csv'):
    user,item,score,days=line.strip().split(',')
    train.setdefault(user,{})
    train[user][item] = score
```

```python
C=dict()
N=dict()
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i] += 1
        C.setdefault(i,{})
        for j in items.keys():
            if i == j : continue
            C[i].setdefault(j,0)
            C[i][j] += 1  #/math.log(1+len(items)*1.0)
```

```python
W = dict()
W_max=dict()
for i,related_items in C.items():
    W.setdefault(i,{})
    for j,cij in related_items.items():
        W_max.setdefault(j,0.0)
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
        if W[i][j]>W_max[j]:
            W_max[j]=W[i][j]
```

```python
for i,related_items in C.items():
    for j,cij in related_items.items():
        W[i][j]=W[i][j]/W_max[j]
```

```python
C=dict()
N=dict()
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i] += 1
        C.setdefault(i,{})
        for j in items.keys():
            if i == j : continue
            C[i].setdefault(j,0)
            C[i][j] += 1
W = dict()
for i,related_items in C.items():
    W.setdefault(i,{})
    for j,cij in related_items.items():
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
```

```python
def Recommend(user):
    rank = dict()
    action_item = train[user]     #用户user产生过行为的item和评分
    for item,score in action_item.items():
        for j,wj in sorted(W[item].items(),key=lambda x:x[1],reverse=True)[0:100]:
            if j in action_item.keys():
                continue
            rank.setdefault(j,0)
            item=int(item)
            days=data0.loc[data0['essay_id']==item].days.unique()[0]
            index=math.exp(-0.4*days)
            rank[j] += float(score) * wj*index
    return dict(sorted(rank.items(),key=lambda x:x[1],reverse=True)[0:50])
```

```python
file=open(r'C:\Users\JDD\query-impala-312176_3.csv','r')
future1={}
for line in file.readlines()[1:]:
    line=line.strip().split(',')
    if not line[0] in future1.keys():
        future1[line[0]] = {line[1]:line[2]}
    else:
        future1[line[0]][line[1]] = line[2]
future1
```

```python
positive=0
devices=[]
for deviceid in future1.keys():
    if deviceid in train.keys():
        devices.append(deviceid)
        positive+=len(future1[deviceid].keys())     
```

```python
positive,len(devices)
```

```python
random.seed(23)
slice=random.sample(devices, 3000)
slice
```

```python
positive=0
for deviceid in slice:
    positive+=len(future1[deviceid].keys())
positive
```

```python
tp=0
true=0
for user in slice:
    recommendations=Recommend(user)
    true+=len(recommendations)
    for item in recommendations.keys():
        if item in future1[user].keys():
            tp+=1
precision=tp/true
recall=tp/positive
F1=2*precision*recall/(precision+recall)
print(tp)
print(true)
print(precision,recall,F1)
```
