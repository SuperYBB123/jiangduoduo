```python
action_diff=dict()
essay_diff=dict()
for line in open(r'C:\Users\JDD\行为得分.csv'):
    user,item,score,actiondays,essaydays=line.strip().split(',')
    action_diff.setdefault(user,{})
    essay_diff.setdefault(user,{})
    action_diff[user][item]=actiondays
    essay_diff[user][item]=essaydays
```

```python
C=dict()
N=dict()
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i] += 1
        C.setdefault(i,{})
        for j in items.keys():
            if i == j : continue
            C[i].setdefault(j,0)
            C[i][j] += 1
W = dict()
W_max=dict()
for i,related_items in C.items():
    W.setdefault(i,{})
    W_max.setdefault(i,0.0)
    for j,cij in related_items.items():
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
        if W[i][j]>W_max[i]:
            W_max[i]=W[i][j]
for i,related_items in C.items():
    for j,cij in related_items.items():
        W[i][j]=W[i][j]/W_max[i]
```

```python
C=dict()
N=dict()
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i] += 1
        C.setdefault(i,{})
        for j in items.keys():
            if i == j : continue
            C[i].setdefault(j,0)
            C[i][j] += 1
W = dict()
for i,related_items in C.items():
    W.setdefault(i,{})
    for j,cij in related_items.items():
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
```

```python
C=dict()
N=dict()
alpha=1
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i] += 1
        C.setdefault(i,{})
        for j in items.keys():
            if i == j : continue
            C[i].setdefault(j,0)
            action1=int(action_diff[user][i])
            action2=int(action_diff[user][j])
            actiondays=abs(action1-action2)
            C[i][j] += 1/(1+alpha*actiondays)
W = dict()
for i,related_items in C.items():
    W.setdefault(i,{})
    for j,cij in related_items.items():
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
```

```python
def Recommend(user):
    rank = dict()
    action_item = train[user]     #用户user产生过行为的item和评分
    for item,score in action_item.items():
        for j,wj in sorted(W[item].items(),key=lambda x:x[1],reverse=True)[0:100]:
            if j in action_item.keys():
                continue
            rank.setdefault(j,0)
            actiondays=int(action_diff[user][item])-14
            essaydays=int(essay_diff[user][item])
            #actionindex=math.exp(-0.4*actiondays)
            essayindex=math.exp(-0.1*essaydays)
            actionindex=1-actiondays/14
            rank[j] += float(score) * wj*actionindex*essayindex
    return dict(sorted(rank.items(),key=lambda x:x[1],reverse=True)[0:50])
```
