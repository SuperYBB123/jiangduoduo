```python
from gensim.models import word2vec
sentences=word2vec.Text8Corpus(r'C:\Users\JDD\Desktop\1.txt')
model=word2vec.Word2Vec(sentences, size = 50, min_count = 20000, window = 5, compute_loss=True,
                  workers = 10, sg=0, sample=1e-4, hs = 0, negative = 40, iter= 1, sorted_vocab = 1)
```

```python
print('')
essay_list=list(model.wv.vocab.keys())
print('vocab',len(essay_list))
essay_sim_degree=dict()
for essay in essay_list:  
    essay_sim_degree.setdefault(essay,{})
    essay_sim_degree[essay][essay]=1
    for x in model.wv.most_similar(essay,topn=50):
        sim=x[0]
        degree=x[1]
        if degree>=0.7:
            essay_sim_degree[essay][sim]=degree
#output=[]
#for e1,x in essay_sim_degree.items():
#    for e2 in x.keys():
#        if e1 in title_dict.keys() and e2 in title_dict.keys() and finish_dict[e2]>0.1:
            #L.append((x[e2],title_dict[e1],title_dict[e2]))
#            output.append([e1,e2,x[e2]])
#output = pd.DataFrame(L, columns=['sim','essay1_title','essay2_title'])
#output = pd.DataFrame(L, columns=['essay1','essay2','sim'])
print('!')
print('',len(essay_sim_degree.keys()))
#print('',len(output))
```
