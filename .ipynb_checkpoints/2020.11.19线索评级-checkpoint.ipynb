```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import recall_score  
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
import warnings
warnings.filterwarnings("ignore")
pd.set_option('max_rows', 100)
```

```python
if_active_today=pd.read_csv(r'C:\Users\JDD\Downloads\当日活跃.csv')
active=pd.read_csv(r'C:\Users\JDD\Downloads\11.24当天是否活跃.csv')
user_active=pd.read_csv(r'C:\Users\JDD\Downloads\11.26是否活跃.csv')
```

```python
#经销商给用户打电话
pos_7=pd.read_csv(r'C:\Users\JDD\Downloads\11.12线索评级正样本7.csv',encoding='gbk')
pos_30=pd.read_csv(r'C:\Users\JDD\Downloads\11.12线索评级正样本30.csv',encoding='gbk')
pos_7.drop(columns=['iforder','active_days','register_days','register_channel','p_day','call_cost'],inplace=True)
pos=pd.merge(pos_30,pos_7,on='uuid',how='left').fillna(0)
#pos=pos_7
pos=pd.merge(pos,active,on='uuid',how='left').fillna(0)
pos['channel']=pos['register_channel'].apply(lambda x:1 if 'AppStore' in x or 'ios' in x else 0)
pos['iforder']=pos['call_cost'].apply(lambda x:1 if x>=120 else 2 if x <=30 else 0)
pos.drop(columns=['uuid','register_channel','call_cost'],inplace=True)
pos_test=pos.loc[(pos['p_day']>=20201123)&(pos['iforder']<=1)]
pos_train=pos.loc[(pos['p_day']<20201123)&(pos['iforder']<=1)]
pos_test.drop(columns=['p_day'],inplace=True)
pos_train.drop(columns=['p_day'],inplace=True)


#用户给经销商打电话
plus_7=pd.read_csv(r'C:\Users\JDD\Downloads\11.26正样本7天.csv',encoding='gbk')
plus_30=pd.read_csv(r'C:\Users\JDD\Downloads\11.26正样本30天.csv',encoding='gbk')
plus_7.drop(columns=['iforder','active_days','register_days','register_channel','p_day','call_cost'],inplace=True)
plus=pd.merge(plus_30,plus_7,on='uuid',how='left').fillna(0)
#plus=plus_7
plus=pd.merge(plus,user_active,on='uuid',how='left').fillna(0)
plus['channel']=plus['register_channel'].apply(lambda x:1 if 'AppStore' in x or 'ios' in x else 0)                                      
plus['iforder']=plus['call_cost'].apply(lambda x:1 if x>=120 else 0)
plus.drop(columns=['uuid','register_channel','call_cost'],inplace=True)
plus_test=plus.loc[plus['p_day']>=20201123]
plus_train=plus.loc[(plus['p_day']<20201123)&(plus['iforder']==1)].sample(200)
plus_test.drop(columns=['p_day'],inplace=True)
plus_train.drop(columns=['p_day'],inplace=True)


#合并两个正样本
p_train=pd.concat([pos_train,plus_train])
```

```python
positive_7_1=pd.read_csv(r'C:\Users\JDD\Downloads\正样本7天1.csv',encoding='gbk')
positive_7_2=pd.read_csv(r'C:\Users\JDD\Downloads\正样本7天2.csv',encoding='gbk')
positive_7_3=pd.read_csv(r'C:\Users\JDD\Downloads\正样本7天3.csv',encoding='gbk')
positive_30_1=pd.read_csv(r'C:\Users\JDD\Downloads\正样本30天1.csv',encoding='gbk')
positive_30_2=pd.read_csv(r'C:\Users\JDD\Downloads\正样本30天2.csv',encoding='gbk')
positive_30_3=pd.read_csv(r'C:\Users\JDD\Downloads\正样本30天3.csv',encoding='gbk')
positive_7=pd.merge(pd.merge(positive_7_1,positive_7_2,on='uuid',how='left'),positive_7_3,on='uuid',how='left')
positive_30=pd.merge(pd.merge(positive_30_1,positive_30_2,on='uuid',how='left'),positive_30_3,on='uuid',how='left')
positive_7.drop(columns=['iforder','active_days','register_days','register_channel','p_day'],inplace=True)
positive=pd.merge(positive_30,positive_7,on='uuid',how='left').fillna(0)
call_cost=pd.read_csv(r'C:\Users\JDD\Downloads\call_cost.csv')
positive=pd.merge(positive,call_cost,on='uuid',how='left')
positive['iforder']=positive['call_cost'].apply(lambda x:1 if x>=90 else 0)
positive.drop(columns=['call_cost'],inplace=True)
positive=pd.merge(positive,if_active_today,on='uuid',how='left').fillna(0)
positive['channel']=positive['register_channel'].apply(lambda x:1 if 'AppStore' in x or 'ios' in x 
                                                         else 0)
positive.drop(columns=['uuid','register_channel'],inplace=True)
positive.rename(columns={'newcar_clickdays_max_30':'newcar_clickdays_max_x', 'newcar_clickdays_max_7':'newcar_clickdays_max_y'},inplace=True)
positive_test=positive.loc[positive['p_day']==20201118]
positive_train=positive.loc[positive['p_day']<20201118]
positive_test.drop(columns=['p_day'],inplace=True)
positive_train.drop(columns=['p_day'],inplace=True)
```

```python
negative_7_1=pd.read_csv(r'C:\Users\JDD\Downloads\负样本7天1.csv')
negative_7_2=pd.read_csv(r'C:\Users\JDD\Downloads\负样本7天2.csv')
negative_7_3=pd.read_csv(r'C:\Users\JDD\Downloads\负样本7天3.csv')
negative_30_1=pd.read_csv(r'C:\Users\JDD\Downloads\负样本30天1.csv')
negative_30_2=pd.read_csv(r'C:\Users\JDD\Downloads\负样本30天2.csv')
negative_30_3=pd.read_csv(r'C:\Users\JDD\Downloads\负样本30天3.csv')
negative_7=pd.merge(pd.merge(negative_7_1,negative_7_2,on='uuid',how='left'),negative_7_3,on='uuid',how='left')
negative_30=pd.merge(pd.merge(negative_30_1,negative_30_2,on='uuid',how='left'),negative_30_3,on='uuid',how='left')
negative_7.drop(columns=['iforder','active_days','register_days','register_channel','p_day'],inplace=True)
negative=pd.merge(negative_30,negative_7,on='uuid',how='left').fillna(0)
negative=pd.merge(negative,if_active_today,on='uuid',how='left').fillna(0)
negative['channel']=negative['register_channel'].apply(lambda x:1 if 'AppStore' in x or 'ios' in x 
                                                         else 0)
negative.drop(columns=['uuid','register_channel','p_day'],inplace=True)
negative.rename(columns={'newcar_clickdays_max_30':'newcar_clickdays_max_x', 'newcar_clickdays_max_7':'newcar_clickdays_max_y'},inplace=True)
```

```python
quantile_1=positive['detail_days_x'].quantile(0.25)
quantile_2=positive['detail_days_x'].quantile(0.5)
quantile_3=positive['detail_days_x'].quantile(0.75)
neg_1=negative.loc[negative['detail_days_x']<=quantile_1].sample(750)
neg_2=negative.loc[(negative['detail_days_x']<=quantile_2)&(negative['detail_days_x']>quantile_1)].sample(750)
neg_3=negative.loc[(negative['detail_days_x']<=quantile_3)&(negative['detail_days_x']>quantile_2)].sample(750)
neg_4=negative.loc[negative['detail_days_x']>quantile_3].sample(750)
neg_sample=pd.concat([neg_1,neg_2,neg_3,neg_4])[pos_train.columns]
#neg_sample=neg_4[pos_train.columns]
data=pd.concat([pos_train,neg_sample])
```

```python
#测试集
predict_data=pos_test
y_predict=predict_data.iforder
x_predict=predict_data.drop(['iforder'], axis=1)
feature_names = list(x_predict.columns)
#训练集和验证集
train_all_data=pos_train
y_train_all=train_all_data.iforder
x_train_all=train_all_data.drop(['iforder'], axis=1)
x_train,x_test,y_train,y_test=train_test_split(x_train_all,y_train_all,test_size=0.2,random_state=23)
#打包dataset
cate_list=['channel']
train_data=lgb.Dataset(data=x_train, label=y_train, feature_name=feature_names,categorical_feature=cate_list)
test_data=lgb.Dataset(data=x_test,label=y_test,feature_name=feature_names,categorical_feature=cate_list)
#模型参数
params = {
    'boosting_type': 'gbdt',  # 设置提升类型
    'objective': 'binary',  # 目标函数
    'metric': {'binary_logloss', 'auc'},  # 评估函数
    'num_trees': 100,
    'max_depth': 8,
    'num_leaves': 128,  # 叶子节点数
    'learning_rate': 0.1,  # 学习速率
    'feature_fraction': 0.8,  # 建树的特征选择比例
    'bagging_fraction': 0.8,  # 建树的样本采样比例
    'bagging_freq': 5,  # 每 k 次迭代执行bagging
    'min_data_in_leaf': 700,
    'max_bin': 256,
    'verbose': 1,  # <0 显示致命的, =0 显示错误 (警告), >0 显示信息
    'is_unbalance' : True,
    'min_data_in_bin': 50
}
num_round=10
#拟合模型
lgb_1=lgb.train(params,train_data,num_round,valid_sets=[test_data])
#各数据集样本数
print('训练集正样本数：'+str(y_train.sum()))
print('训练集负样本数：'+str(len(x_train)-y_train.sum()))
print('测试集正样本数：'+str(y_predict.sum()))
print('测试集负样本数：'+str(len(x_predict)-y_predict.sum()))
```

```python
#预测与模型评估
y_pred=lgb_1.predict(x_predict)
y_pred_binary=np.where(y_pred>0.5,1,0)
acc=accuracy_score(y_predict,y_pred_binary)
recall=recall_score(y_predict,y_pred_binary)
auc=roc_auc_score(y_predict,y_pred)
print('auc:',auc,'acc:',acc,'recall',recall)
```

```python
#特征重要性可视化
import matplotlib.pylab as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['FangSong'] #中文乱码

feature_import=list(lgb_1.feature_importance(importance_type = 'gain'))
#feature_names.extend(['detail_price_consult_z','detail_clicks_z','detail_clicktimes_max_z'])
#feature_import.extend([510.34725668968,148.34678832837,110.34643778])
feature_names_pd = pd.DataFrame({'column': feature_names,
                                 'importance': feature_import,
                                 })
feature_names_pd=feature_names_pd.sort_values(by='importance',ascending=False)
print(feature_names_pd[1:15])
cols = feature_names_pd[["column", "importance"]].groupby("column").mean().sort_values(by="importance",ascending=False)[1:30].index

best_features = feature_names_pd.loc[feature_names_pd.column.isin(cols)]

plt.figure(figsize=(10, 15))
sns.barplot(x="importance", y="column", data=best_features.sort_values(by="importance", ascending=False))
plt.title('LightGBM Features')
plt.tight_layout()
```
