```python
#读取7天数据
data7=pd.read_csv(r'C:\Users\JDD\Downloads\query-impala-311314.csv')
data7.drop(['deviceid','essay_id.1','c_title','d_dateline'],axis=1,inplace=True)
```

```python
#获取文章id
essay_list=list(data7['essay_id'].unique())
essay_list
```

```python
#获取文章热度
essay_heat={}
for essay in essay_list:
    clicks=data7.loc[data7['essay_id']==essay].clicktimes.sum()
    exposes=data7.loc[data7['essay_id']==essay].exposuretimes.sum()
    heat=clicks/exposes
    if clicks>=500:
        essay_heat.update({essay:heat})
essay_sorted=sorted(essay_heat.items(),key=lambda i:i[1],reverse=True)
```

```python
#根据热度获得全局热度推荐列表
Heat_recommendations=[]
for i in range(50):
    Heat_recommendations.append(essay_sorted[i][0])
heat_recommendations=[]
for item in Heat_recommendations:
    heat_recommendations.append(str(item))
heat_recommendations 
```

```python
#userCF
#计算任意两用户的jaccard距离作为相似度
def jaccard(user1,user2):
    user1_data=Click_data[user1]
    user2_data=Click_data[user2]
    common=[]
    for essay in user1_data.keys():
        if essay in user2_data.keys():
            common.append(essay)
    if len(common) == 0:
        return 0
    n=len(common)
    sum1=len(user1_data.keys())
    sum2=len(user2_data.keys())
    jaccard=n/math.sqrt(sum1*sum2)
    return jaccard

#获取相似用户列表
def Top_simliar(deviceid):
    res = []
    for userid in Click_data.keys():
        if not userid == deviceid:
            simliar = jaccard(deviceid,userid)
            res.append((userid,simliar))
    res.sort(key=lambda val:val[1],reverse=True)
    return res[:200]

#根据相似用户推荐文章
def recommend(user):                                         #userCF推荐列表
    top_sim_users=[]
    rec=[]
    for res in Top_simliar(user):
        top_sim_users.append(res[0])
    for top_sim_user in top_sim_users:
        items = Click_data[top_sim_user]
        for item in items.keys():
            if item not in Click_data[user].keys() and item not in rec:
                rec.append(item)
    rec1=list(set(rec))
    rec1.sort(key=rec.index)
    return rec1[:50]
```

```python
#itemCF
train=dict()
for line in open(r'C:\Users\JDD\query-impala-310822_7.csv'):
    user,item,score=line.strip().split(',')
    train.setdefault(user,{})
    train[user][item] = score
C=dict()
N=dict()
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i] += 1
        C.setdefault(i,{})
        for j in items.keys():
            if i == j : continue
            C[i].setdefault(j,0)
            C[i][j] += 1
W = dict()
for i,related_items in C.items():
    W.setdefault(i,{})
    for j,cij in related_items.items():
        W[i][j] = cij / (math.sqrt(N[i] * N[j]))
```

```python
def Recommend(user):
    rank = dict()
    action_item = train[user]     #用户user产生过行为的item和评分
    for item,score in action_item.items():
        for j,wj in sorted(W[item].items(),key=lambda x:x[1],reverse=True)[0:200]:
            if j in action_item.keys():
                continue
            rank.setdefault(j,0)
            rank[j] += float(score) * wj
    return dict(sorted(rank.items(),key=lambda x:x[1],reverse=True)[0:200]).keys()
```
