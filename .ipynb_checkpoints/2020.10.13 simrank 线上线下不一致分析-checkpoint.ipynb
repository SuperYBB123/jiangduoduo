```python
import pandas as pd
from gensim.models import word2vec
import math
import random
import numpy as np
```

```python
tmp=pd.read_csv(r'C:\Users\JDD\Downloads\train.csv')
fl = open(r'F:\work\5.18 word2vec训练文章id\data1.txt', 'w', encoding='utf8')
for n in range(20):
    for i in tmp.index:
        j=tmp['list'][i]
        k=j.strip().split(',')
        if len(k)>=5:
            random.shuffle(k)
            for wd in k:
                fl.write(wd)
                fl.write(' ')
            fl.write('\n')
    print(n+1)
fl.close()
```

```python
sentences=word2vec.Text8Corpus(r'F:\work\5.18 word2vec训练文章id\data1.txt')
model=word2vec.Word2Vec(sentences,size=50,min_count=800,window=5,compute_loss=True,
                        workers=10,sg=0,sample=1e-4,hs=0,negative=40,iter= 1,sorted_vocab=1)
#sentences,sg=1,hs=1,min_count=1,size=25,negative=15,sample=1e-3,ns_exponent=0.75,window=3,iter=1
#model.save(r'F:\work\5.18 word2vec训练文章id\model\w2v_shuffle_skip_5')
```

```python
essay_list=list(model.wv.vocab.keys())
essay_sim_degree=dict()
for essay in essay_list:
    essay_sim_degree.setdefault(essay,{})
    for x in model.wv.most_similar(essay,topn=50):
        sim=x[0]
        degree=x[1]
        if degree>0.7:
            essay_sim_degree[essay][sim]=degree
len(essay_sim_degree.keys())
```

```python
scan_dict={}
for i in tmp.index:
    deviceid=tmp['deviceid'][i]
    scan_dict.setdefault(deviceid,{})
    j=tmp['list'][i]
    k=j.strip().split(',')
    goal=k[-5:]
    for item in goal:
        score=5-int(goal.index(item))
        scan_dict[deviceid][item]=score
scan_dict
```
