```python
rec=dict()
read=dict()
with open(r'F:\work\4.8 热门推荐\7.00+4_1.csv',encoding='utf_8_sig') as f:
    f.readline()
    for line in f:
        user,item,score=line.strip().split(',')
        rec.setdefault(user,{})
        if score=='1':
            read.setdefault(user,{})
            read[user][item]=score
        else:
            rec[user][item]=score
```

```python
rec
```

```python
read
```

```python
devices=[]
for user in read.keys():
    if user in rec.keys():
        devices.append(user)
```

```python
right=0
predict=0
Read=0
for user in devices:
    recommend=hotall
    rec_un=[]
    for item in recommend:
        if str(item) in rec[user]:
            rec_un.append(item)
    if len(rec_un)>0:
        predict+=len(rec_un)
        Read+=len(read[user])
        for item in rec_un:
            if str(item) in read[user].keys():
                right+=1
precision=right/predict
recall=right/Read
print('top50recall=',recall)
print('top50precision=',precision)
```
