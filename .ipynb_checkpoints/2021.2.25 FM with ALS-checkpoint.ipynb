```python
import pandas as pd
import numpy as np
import math
import random
import implicit
from sklearn import preprocessing
from scipy.sparse import coo_matrix, csr_matrix
from implicit.als import AlternatingLeastSquares
import tensorflow as tf
from tqdm import trange
```

```python
def get_items(data):
    items=data.split(',')
    return len(items)
```

```python
data=pd.read_csv(r'F:\work\Downloads\MF半月点击行为.csv')
data['item_list']=data['list'].apply(lambda x:get_items(x))
data1=data.loc[data['item_list']>5]
data1['list']=data1['list'].map(lambda x:x.split(','))
data1=data1[['deviceid','list']]
data2=data1.explode('list')
data2['ratings']=1
data2=data2.reset_index(drop=True)
data2
```

```python
le = preprocessing.LabelEncoder()
data2['ratings']=1
user_id = data2["deviceid"].values
user_lst=np.unique(user_id)
le.fit(user_lst)
user_id_df=pd.DataFrame(user_lst,columns=['deviceid'])
user_id=le.transform(user_id)
item_id = data2["list"].values
le = preprocessing.LabelEncoder()
item_lst=np.unique(item_id)
le.fit(item_lst)
item_id_df=pd.DataFrame(item_lst,columns=['itemid'])
item_id=le.transform(item_id)
ratings = data2["ratings"].values
item_user_data = csr_matrix((ratings,(item_id,user_id)),shape=(max(item_id) + 1, max(user_id) + 1))
```

# 数据处理

```python
data=pd.read_csv(r'F:\work\Downloads\FM2-22.csv')
pos_data=data.loc[data['action_type']==1]
u_data=pos_data[['deviceid','entrance_code','n_activate_utilnow','province','plateform']]
i_data=data[['itemid','label','d_st_ctime','d_st_contenttype']]
u_data.drop_duplicates(inplace=True)
u_data=u_data.dropna()
i_data.drop_duplicates(inplace=True)
```

```python
#user侧特征
u_data=pd.merge(u_data,user_df,how='inner',on='deviceid')

#提取entrance
def get_entrance(data):
    data=data.split('+')[0]
    data=data.split('--')[0]
    data=data.split('%')[0]
    return data
#province取前两位
def get_province(data):
    return int(str(data)[:2])


u_data['entrance']=u_data['entrance_code'].apply(lambda x:get_entrance(x))
entrances=pd.get_dummies(u_data['entrance'],prefix='entrance')
u_data=pd.concat([u_data,entrances],axis=1)

u_data['province']=u_data['province'].apply(lambda x:get_province(x))
provinces=pd.get_dummies(u_data['province'],prefix='province')
u_data=pd.concat([u_data,provinces],axis=1)

#提取plateform
plateforms=pd.get_dummies(u_data['plateform'],prefix='plateform')
u_data=pd.concat([u_data,plateforms],axis=1)

#连续型特征分桶
u_data['activate_time']=u_data['n_activate_utilnow'].apply(lambda x:1 if x<=1 
                                                               else 2 if x>1 and x<=7
                                                              else 3 if x>7 and x<=15
                                                              else 4 if x>15 and x<=30
                                                              else 5 if x>30 and x<=45
                                                              else 6 if x>45 and x<=60
                                                              else 7)
activate_times=pd.get_dummies(u_data['activate_time'],prefix='activate_time')
u_data=pd.concat([u_data,activate_times],axis=1)
```

```python
#item侧特征
i_data=pd.merge(i_data,item_df,how='inner',on='itemid')

#获取所有标签
label_list=[]
for i in i_data['label'].unique():
    j=i.split(',')
    for k in j:
        if k not in (label_list) and k!='暂无':
            label_list.append(k)
for i in label_list:
    column_name='label_'+i
    i_data[column_name]=i_data['label'].apply(lambda x:1 if i in x else 0)


#提取文章类别
contenttypes=pd.get_dummies(i_data['d_st_contenttype'],prefix='contenttype')
i_data=pd.concat([i_data,contenttypes],axis=1)

i_data['st_time']=i_data['d_st_ctime'].apply(lambda x:1 if x<=86400 
                                                           else 2 if x>86400 and x<=172800
                                                          else 3 if x>172800 and x<=345600
                                                          else 4 if x>345600 and x<=518400
                                                          else 5 if x>518400 and x<=604800
                                                          else 6)
st_times=pd.get_dummies(i_data['st_time'],prefix='st_time')
i_data=pd.concat([i_data,st_times],axis=1)
```

```python
#获取user、item、context侧特征集合
u_data=u_data.drop(columns=['entrance_code','n_activate_utilnow','province','plateform','entrance','activate_time'])
i_data=i_data.drop(columns=['label','d_st_ctime','d_st_contenttype','st_time'])
user_col=[]
item_col=[]
context_col=[]
user_pre=['activate_time','entrance','user_emb']
item_pre=['label','st_time','contenttype','item_emb']
context_pre=['province','plateform']
for i in u_data.columns:
    for j in user_pre:
        if j in i:
            user_col.append(i)
    for j in context_pre:
        if j in i:
            context_col.append(i)
for i in i_data.columns:
    for j in item_pre:
        if j in i:
            item_col.append(i)
```

# FM

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Layer,Dense,Input
from tensorflow.keras import Model,activations
from tensorflow.keras import backend as K
```

```python
tf.keras.backend.clear_session()
class MyLayer(tf.keras.layers.Layer):
    def __init__(self, input_dim, output_dim = 100, **kwargs):
        self.input_dim = input_dim
        self.output_dim = output_dim
        super(MyLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        self.kernel = self.add_weight(name='kernel', 
                                      shape=(self.input_dim, self.output_dim),
                                      initializer='glorot_uniform',
                                      trainable=True)
        super(MyLayer, self).build(input_shape)

    def call(self, x):
        a = K.pow(K.dot(x,self.kernel), 2)
        b = K.dot(K.pow(x, 2), K.pow(self.kernel, 2))
        return K.mean(a-b, 1, keepdims=True)*0.5

    def compute_output_shape(self, input_shape):
        return (input_shape[0], self.output_dim)
```

```python
def FM(feature_dim, metrics):
    inputs = Input(shape=(feature_dim,))
    #cnt_negative=16
    #negatives = Lambda(lambda x: K.random_uniform((K.shape(x)[0], cnt_negative), 0, feature_dim, 'int32'))(inputs)
    #samples = Lambda(lambda x: K.concatenate(x))([inputs,negatives])
    #input = {}
    #for f in feature_names:
    #    input[f] = Input(shape=(1,), name = f, dtype = x_train[f].dtype)
    #generate feature_columns
    #feature_layer = tf.keras.layers.DenseFeatures(feature_columns)
    #Transform input 
    #dense_feature = feature_layer(input)
#    liner = tf.keras.layers.Dropout(0.25)(dense_feature)
    #线性部分(w*x + b)
    liner = tf.keras.layers.Dense(units=1, 
#                                  bias_regularizer=tf.keras.regularizers.l2(0.01),
                                  kernel_regularizer=tf.keras.regularizers.l1(0.02),
                                  )(inputs)
#    cross = tf.keras.layers.Dropout(0.25)(dense_feature)
    #交叉项
    cross = MyLayer(feature_dim)(inputs)
    add = tf.keras.layers.Add()([liner, cross])
    predictions = tf.keras.layers.Activation('sigmoid')(add)
    
    model = tf.keras.Model(inputs=inputs, outputs=predictions)
    model.compile(loss='binary_crossentropy',
                  optimizer=tf.keras.optimizers.Adam(lr=0.0001),
                  metrics=metrics)
    return model
```

```python
def train_fm(feature_dim):
    metrics = [
    tf.keras.metrics.BinaryAccuracy(name='accuracy'),
    tf.keras.metrics.AUC(name='auc')]
    
    fm = FM(feature_dim, metrics)
    fm.fit(x_train,y_train,
           batch_size=128,
            validation_data=(x_test,y_test),
          epochs=20,verbose=2) #, class_weight = 'auto'
    return fm
```

```python
input_dim=x_train.shape[1]
fm = train_fm(input_dim)
```

```python
V=fm.get_weights()[2]
len_user=len(user_col)
len_item=len(item_col)
len_context=len(context_col)
V_user=V[:len_user]
V_item=V[len_user:len_user+len_item]
V_context=V[len_user+len_item:len_user+len_item+len_context]


#计算user、item、context对应的embedding
def get_part_embedding(feature,emb):
    avg_emb=np.zeros((1,emb.shape[1]))
    for i in range(len(feature)):
        avg_emb+=feature[i]*emb[i]
    avg_emb=avg_emb/len(feature)
    return avg_emb
```
