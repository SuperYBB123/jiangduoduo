```python
import urllib.request as urlrequest
weather_url='http://www.weather.com.cn/weather/101190401.shtml'
web_page=urlrequest.urlopen(weather_url).read()
print(web_page)
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(web_page,'html.parser')
soup.find(id='7d').get_text()
```

```python
forecast_text=soup.find(id='7d').get_text()
forecast_text.split('\n')
```

```python
soup_forecast=soup.find(id='7d')
weather_list=soup_forecast.find_all(class_='wea')
for i in range(7):
    weather=weather_list[i].get_text()
    print(weather)
```
