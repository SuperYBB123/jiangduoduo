```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\JDD\Downloads\20191121.csv')
df
```

```python
df_2=pd.read_csv(r'C:\Users\JDD\Downloads\20191205.csv')
df_2
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['ctr2'],df['action2'],c='blue')
ax.scatter(df_2['ctr2'],df_2['action2'],c='red')
plt.show()
```

```python
essay_list=list(df['essay_id'].unique())
essay_list
```

```python
std=np.std(df['ctr2'])
std2=np.std(df['action2'])
mean=np.mean(df['ctr2'])
mean2=np.mean(df['action2'])
for essay in essay_list:
    index=df.loc[df['essay_id']==essay].index
    ctr_std=float(df.loc[df['essay_id']==essay].ctr_std)
    action_std=float(df.loc[df['essay_id']==essay].action_std)
    if abs(ctr_std)>3*std:
        if ctr_std>0:
            df.loc[index,'ctr2']=3*std+mean
        else:
            df.loc[index,'ctr2']=-3*std+mean
    if abs(action_std)>3*std2:
        if action_std>0:
            df.loc[index,'action2']=3*std2+mean2
        else:
            df.loc[index,'action2']=-3*std2+mean2
```

```python
std=np.std(df_2['ctr2'])
std2=np.std(df_2['action2'])
mean=np.mean(df_2['ctr2'])
mean2=np.mean(df_2['action2'])
for essay in essay_list:
    index=df_2.loc[df_2['essay_id']==essay].index
    ctr_std=float(df_2.loc[df_2['essay_id']==essay].ctr_std)
    action_std=float(df_2.loc[df_2['essay_id']==essay].action_std)
    if abs(ctr_std)>3*std:
        if ctr_std>0:
            df_2.loc[index,'ctr2']=3*std+mean
        else:
            df_2.loc[index,'ctr2']=-3*std+mean
    if abs(action_std)>3*std2:
        if action_std>0:
            df_2.loc[index,'action2']=3*std2+mean2
        else:
            df_2.loc[index,'action2']=-3*std2+mean2
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(df['ctr2'],df['action2'],c='blue')
ax.scatter(df_2['ctr2'],df_2['action2'],c='red')
plt.show()
```

```python
auther_list=[]
for auther in df_2['n_autherid'].unique():
    auther_list.append(auther)
auther_list
```

```python
output=pd.DataFrame(columns=['auther','avg_ctr','avg_action'])
for auther in auther_list:
    length=len(df.loc[df['n_autherid']==auther])
    sum_ctr=sum(df.loc[df['n_autherid']==auther].ctr2)
    sum_action=sum(df.loc[df['n_autherid']==auther].action2)
    avg_ctr=sum_ctr/length
    avg_action=sum_action/length
    new=pd.DataFrame({'auther':auther,'avg_ctr':avg_ctr,'avg_action':avg_action},index=[1])
    output=output.append(new,ignore_index=True)
output
```

```python
output.to_csv(r'C:\Users\JDD\Desktop\20191205标准化后的ctr、action.csv',sep=',',index=False,header=True)
```
