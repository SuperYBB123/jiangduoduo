```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\JDD\Downloads\1107device+p_day.csv')
df
```

```python
car_list=list(df['car_id'].unique())
len(car_list)
```

```python
output=pd.DataFrame(columns=['car_id','device_cnt','essay_cnt'])
for car in car_list:
    device_cnt=len(df.loc[df['car_id']==car].deviceid.unique())
    essay_cnt=len(df.loc[df['car_id']==car].essay_id.unique())
    new=pd.DataFrame({'car_id':car,'device_cnt':device_cnt,'essay_cnt':essay_cnt},index=[1])
    output=output.append(new,ignore_index=True)
output
```

```python
output.to_csv(r'C:\Users\JDD\Desktop\1107device-carID作者数文章数.csv',sep=',',index=False,header=True)
```
