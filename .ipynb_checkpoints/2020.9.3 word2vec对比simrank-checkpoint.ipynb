```python
import pandas as pd
from gensim.models import word2vec
import math
import random
import numpy as np
```

```python
tmp=pd.read_csv(r'C:\Users\JDD\Downloads\11107天_2.csv')
fl = open(r'F:\work\5.18 word2vec训练文章id\data1.txt', 'w', encoding='utf8')
for n in range(20):
    for i in tmp.index:
        j=tmp['list'][i]
        k=j.strip().split(',')
        if len(k)>=5:
            random.shuffle(k)
            for wd in k:
                fl.write(wd)
                fl.write(' ')
            fl.write('\n')
    print(n+1)
fl.close()
```

```python
sentences=word2vec.Text8Corpus(r'F:\work\5.18 word2vec训练文章id\data1.txt')
model=word2vec.Word2Vec(sentences, size = 50, min_count = 2000, window = 5, compute_loss=True,
                  workers = 10, sg=0, sample=1e-4, hs = 0, negative = 40, iter= 1, sorted_vocab = 1)
#sentences,sg=1,hs=1,min_count=1,size=25,negative=15,sample=1e-3,ns_exponent=0.75,window=3,iter=1
#model.save(r'F:\work\5.18 word2vec训练文章id\model\w2v_shuffle_skip_5')
```

```python
df=pd.read_csv(r'C:\Users\JDD\Downloads\1110title.csv')
df=df.loc[df['c_title'].notna()]
title_dict=dict()
for i in df.index:
    itemid=df['itemid'][i]
    title=df['c_title'][i]
    title_dict[str(itemid)]=title
```

```python
essay_list=list(model.wv.vocab.keys())
essay_sim_degree=dict()
for essay in essay_list:
    if essay in title_dict.keys():
        essay_sim_degree.setdefault(essay,{})
        for x in model.wv.most_similar(essay,topn=50):
            sim=x[0]
            degree=x[1]
            if sim in title_dict.keys():
                essay_sim_degree[essay][sim]=degree
```

```python
essay_list=list(model.wv.vocab.keys())
essay_sim_degree_1=dict()
for essay in essay_list:
    if essay in title_dict.keys():
        essay_sim_degree_1.setdefault(essay,{})
        for x in model.wv.most_similar(essay,topn=50):
            sim=x[0]
            degree=x[1]
            if sim in title_dict.keys():
                essay_sim_degree_1[essay][sim]=degree
```

```python
train={}
tmp=pd.read_csv(r'C:\Users\JDD\Downloads\1月观看.csv')
for i in tmp.index:
    deviceid=tmp['deviceid'][i]
    j=tmp['list'][i].strip().split(',')
    train[deviceid]=j
test={}
df=pd.read_csv(r'F:\work\8.19 simrank\test_曝光.csv')
for i in df.index:
    deviceid=df['deviceid'][i]
    action=df['action_type'][i]
    j=df['list'][i]
    item_list=j.strip().split(',')
    test.setdefault(deviceid,{})
    test[deviceid].setdefault(0,[])
    test[deviceid].setdefault(1,[])
    test[deviceid][action]=item_list
```

```python
test
```

```python
device_list=[x for x in train.keys() if x in test.keys()]
sample_device=random.sample(device_list,50000)
PRECISION=[]
RECALL=[]
expose_ACC=[]
for device in sample_device:
    rec_dict={}
    watch_list=train[device]
    for item in watch_list:
        if item in essay_sim_degree_1.keys():
            simdict=essay_sim_degree_1[item]
            for sim_item,sim in simdict.items():
                rec_dict[sim_item]=sim
    rec=sorted(rec_dict, key=lambda x:x[1], reverse=True)[:50]
    expose=[x for x in rec if x in test[device][0]]
    if len(expose)>0 and len(test[device][1])>0:
        precision=len([x for x in rec if x in test[device][1]])/len(rec)
        recall=len([x for x in test[device][1] if x in rec])/len(test[device][1])
        exposed_acc=len([x for x in expose if x in test[device][1]])/len(expose)
        PRECISION.append(precision)
        RECALL.append(recall)
        expose_ACC.append(exposed_acc)
print(np.mean(PRECISION),np.mean(RECALL),np.mean(expose_ACC),len(RECALL))
```
