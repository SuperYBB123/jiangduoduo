```python
rec=dict()
read=dict()
with open(r'F:\work\4.8 热门推荐\0407.csv',encoding='utf_8_sig') as f:
    f.readline()
    for line in f:
        user,item,score=line.strip().split(',')
        rec.setdefault(user,{})
        if score=='1':
            read.setdefault(user,{})
            read[user][item]=score
        else:
            rec[user][item]=score
```

```python
rec
```

```python
devices=[]
for user in rec.keys():
    if user in read.keys():
        devices.append(user)
len(devices)
```

```python
right=0
predict=0
Read=0
for user in devices:
    reclen=len(rec[user])
    n=min(reclen,50)
    predict+=n
    Read+=len(read[user])
    items=list(rec[user].keys())[:n]
    for item in items:
        if item in read[user]:
            right+=1
precision=right/predict
recall=right/Read
print('top50recall=',recall)
print('top500precision=',precision)
```
