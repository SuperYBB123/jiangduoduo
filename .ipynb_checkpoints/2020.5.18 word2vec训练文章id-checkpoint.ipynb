```python
titles=pd.read_csv(r'F:\work\5.18 word2vec训练文章id\essay_title.csv')
titles.itemid = (titles['itemid']).astype(str)
titles.c_title = (titles['c_title']).astype(str)
title_dict=titles.set_index('itemid')['c_title'].to_dict()
title_dict
```

```python
#数据转为字典
train=dict()
with open(r'F:\work\5.18 word2vec训练文章id\data.csv',encoding='utf_8_sig') as f:
    f.readline()
    for line in f:
        user,itemid=line.strip().split(',')
        train.setdefault(user,[])
        train[user].append(itemid)
train
```

```python
L=[(k,v[:-1]) for k, v in train.items()]
data= pd.DataFrame(L, columns=['user','words'])
fl = open(r'F:\work\5.18 word2vec训练文章id\data.txt', 'w', encoding='utf8')
for i in data.index:
    a = list()
    print(i)
    for j in data['words'][i]:
        a.append(j)
    if len(a)>0:
        fl.write(str(a))
        fl.write('\n')
fl.close()
```

```python
sentences=word2vec.Text8Corpus(r'F:\work\5.18 word2vec训练文章id\data.txt')
model=word2vec.Word2Vec(sentences,window=3)
model.save(r'F:\work\5.18 word2vec训练文章id\w2v_essayid')
```

```python
essay_list=list(d['itemid'].unique())
essay_sim=dict()
essay_sim_degree=dict()
for essay in essay_list:
    if str(essay) in model.wv.vocab.keys():
        essay_sim.setdefault(str(essay),{})
        essay_sim_degree.setdefault(str(essay),{})
        for i,x in enumerate(model.most_similar(str(essay),topn=5)):
            sim=x[0]
            rank=i+1
            degree=x[1]
            essay_sim[str(essay)][sim]=rank
            essay_sim_degree[str(essay)][sim]=degree
```

```python
L = [(x[e2],title_dict[e1],title_dict[e2]) for e1,x in essay_sim_degree.items() for e2 in x.keys()]
output = pd.DataFrame(L, columns=['sim','essay1_title','essay2_title'])
output.to_csv(r'F:\work\5.18 word2vec训练文章id\essay_sim.csv',index=False)
```

```python
#icf
train1=dict()
for line in open(r'F:\work\5.18 word2vec训练文章id\data.csv'):
    user,item = line.strip().split(',')
    train1.setdefault(user,{})
    train1[user][item]=str(1)
```

```python
C = dict()  # 文章-文章的共现矩阵
N = dict()  # 文章被多少个不同用户阅读
for user, items in train1.items():
    for i in items.keys():
        N.setdefault(i, 0)
        N[i] += 1
        C.setdefault(i, {})
        for j in items.keys():
            if i == j: continue
            C[i].setdefault(j, 0)
            C[i][j] += 1
```

```python
similarity = dict()  # 相似度矩阵
for i, related_items in C.items():
    similarity.setdefault(i, {})
    for j, cij in related_items.items():
        similarity[i][j] = cij / (math.sqrt(N[i] * N[j]))
for i in similarity.keys():
    if len(similarity[i])>0:
        max_v=max(similarity[i].values())
        for j in similarity[i].keys():
            sim=similarity[i][j]
            similarity[i][j]=sim/max_v
```

```python
L_icf=[]
for e1,x in similarity.items():
    for e2 in x.keys():
        L_icf.append((x[e2],title_dict[e1],title_dict[e2]))
#L = [(x[e2],title_dict[e1],title_dict[e2]) for e1,x in similarity.items() for e2 in x.keys()]
#output_icf = pd.DataFrame(L_icf, columns=['sim','essay1_title','essay2_title'])
#output_icf.to_csv(r'F:\work\5.18 word2vec训练文章id\essay_sim_icf.csv',index=False)
```
