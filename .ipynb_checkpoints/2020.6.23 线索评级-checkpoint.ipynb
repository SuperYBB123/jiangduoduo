```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv(r'F:\work\6.12 线索评级\new\评级.csv',encoding='gbk')
df_area=pd.read_csv(r'F:\work\6.12 线索评级\new\areacode.csv')
df_consult=pd.read_csv(r'F:\work\6.12 线索评级\new\询价成功.csv')
df_first_purchase=pd.read_csv(r'F:\work\6.12 线索评级\new\purchase.csv')
df_content=pd.read_csv(r'F:\work\6.12 线索评级\new\content.csv',encoding='gbk')
df_pic=pd.read_csv(r'F:\work\6.12 线索评级\new\pic.csv')
df
```

```python
df=pd.merge(df,df_area,on='uuid',how='left')
df=pd.merge(df,df_consult,on='uuid',how='left')
df=pd.merge(df,df_first_purchase,on='uuid',how='left')
df=pd.merge(df,df_content,on='uuid',how='left')
df=pd.merge(df,df_pic,on='uuid',how='left')
df
```

```python
df.columns
```

```python
data=df[['uuid', 'iforder','active_days', 'register_days', 'register_channel','areacode','first_purchase_days','detail_clicks', 'detail_distinct', 'detail_days',
       'detail_clicktimes_max', 'detail_config', 'detail_distributor',
       'detail_reputation', 'detail_info', 'detail_ceping', 'detail_follow',
       'detail_compare', 'detail_price_reduction', 'detail_price_consult',
       'detail_consult_days', 'detail_similar', 'newcar_clicks',
       'newcar_distinct', 'newcar_days', 'newcar_clicktimes_max',
       'newcar_clickdays_max', 'newcar_picture', 'newcar_config',
       'newcar_subsidy', 'newcar_comments', 'newcar_customer_service',
       'newcar_store', 'newcar_color', 'newcar_similar',
       'newcar_order_confirm', 'newcar_pay', 'moment_num', 'consult_price_times', 'consult_avg_price',
       'maxpicpv', 'picpvrate','maxpv','essay_num']]
data['iforder'].replace('c','C',inplace=True)
data.to_csv(r'F:\work\6.12 线索评级\new\正负.csv',index=False)
```
