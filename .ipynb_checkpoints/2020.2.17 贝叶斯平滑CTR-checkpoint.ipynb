```python
data7=pd.read_csv(r'C:\Users\JDD\Downloads\query-impala-311314.csv')
data7.drop(['deviceid','essay_id.1','c_title','d_dateline'],axis=1,inplace=True)
```

```python
essay_list=list(data7['essay_id'].unique())
essay_list
```

```python
essay_heat={}
for essay in essay_list:
    clicks=data7.loc[data7['essay_id']==int(essay)].clicktimes.sum()
    exposes=data7.loc[data7['essay_id']==int(essay)].exposuretimes.sum()
    heat=clicks/exposes
    essay_heat.update({essay:heat})
```

```python
mean=np.mean(list(essay_heat.values()))
var=np.var(list(essay_heat.values()))
alpha=mean*(mean*(1-mean)/var-1)
beta=(1-mean)*(mean*(1-mean)/var-1)
print(mean,var,alpha,beta)
```

```python
essay_heat={}
for essay in essay_list:
    clicks=data7.loc[data7['essay_id']==int(essay)].clicktimes.sum()
    exposes=data7.loc[data7['essay_id']==int(essay)].exposuretimes.sum()
    heat=clicks/exposes
    if clicks>=100:
        essay_heat.update({essay:heat})
essay_sorted=sorted(essay_heat.items(),key=lambda i:i[1],reverse=True)
essay_sorted
```

```python
Heat_recommendations=[]
for i in range(50):
    Heat_recommendations.append(essay_sorted[i][0])
heat_recommendations=[]
for item in Heat_recommendations:
    heat_recommendations.append(str(item))
Heat_recommendations
```

```python
Bayes_ctr={}
for essay in essay_list:
    clicks=data7.loc[data7['essay_id']==int(essay)].clicktimes.sum()
    exposes=data7.loc[data7['essay_id']==int(essay)].exposuretimes.sum()
    heat=(clicks+alpha)/(exposes+alpha+beta)
    if clicks>=100:
        Bayes_ctr.update({essay:heat})
bayes_essay_sorted=sorted(Bayes_ctr.items(),key=lambda i:i[1],reverse=True)
bayes_essay_sorted
```

```python
Bayes_Heat_recommendations=[]
for i in range(50):
    Bayes_Heat_recommendations.append(bayes_essay_sorted[i][0])
bayes_heat_recommendations=[]
for item in Bayes_Heat_recommendations:
    bayes_heat_recommendations.append(str(item))
bayes_heat_recommendations
```
