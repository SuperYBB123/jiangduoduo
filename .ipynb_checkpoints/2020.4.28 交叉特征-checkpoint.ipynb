```python
import numpy as np
data=pd.read_csv(r'F:\work\4.20 ctr预估\0418.csv')
```

```python
province_label_cross=pd.DataFrame(columns=['itemid','ctr','device_cnt','cross_label'])
cross=data.groupby('itemid')['action_type'].agg({'click':'sum','ctr':'count'})
cross['expose']=cross['ctr']-cross['click']
a=cross['click'].mean()
b=cross['expose'].mean()
cross['ctr']=(cross['click']+a)/(cross['expose']+b)
cross2=data.groupby(['itemid']).agg({'deviceid':pd.Series.nunique})
cross=pd.merge(cross,cross2,how='inner',on='itemid')
cross=cross.drop(columns=['click','expose'])
cross['cross_label']='1_搞笑'
cross.rename(columns={'deviceid':'device_cnt'},inplace=True)
cross=cross.reset_index()
province_label_cross=pd.concat([province_label_cross,cross],axis=0)
province_label_cross
```
