```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARMA
import warnings
warnings.filterwarnings('ignore')
from itertools import product
from datetime import datetime
import requests
import json
endpoint = 'https://min-api.cryptocompare.com/data/histoday'
res = requests.get(endpoint + '?fsym=BTC&tsym=USD&limit=2000')
hist = pd.DataFrame(json.loads(res.content)['Data'])
hist = hist.set_index('time')
hist.index = pd.to_datetime(hist.index, unit='s')
hist

```

```python
df_month = hist.resample('M').mean()
df_Q = hist.resample('Q-DEC').mean()
df_year = hist.resample('A-DEC').mean()

#æŒ‰å¤©ï¼Œæœˆï¼Œå­£åº¦ï¼Œå¹´æ¥æ˜¾ç¤ºæ¯”ç‰¹å¸çš„èµ°åŠ¿
fig = plt.figure(figsize=[15,7])
plt.rcParams['font.sans-serif'] = ['SimHei']  #ç”¨æ¥æ­£å¸¸æ˜¾ç¤ºä¸­æ–‡æ ‡ç­¾
plt.suptitle('æ¯”ç‰¹å¸é‡‘é¢ï¼ˆç¾Žé‡‘ï¼‰',fontsize = 20) #ç»˜åˆ¶æ€»å›¾çš„æ ‡é¢˜
plt.subplot(221)
plt.plot(hist.open,'-',label = 'æŒ‰å¤©')
plt.legend()
plt.subplot(222)
plt.plot(df_month.open,'-',label = 'æŒ‰æœˆ')
plt.legend()
plt.subplot(223)
plt.plot(df_Q.open,'-',label = 'æŒ‰å­£åº¦')
plt.legend()
plt.subplot(224)
plt.plot(df_year.open,'-',label = 'æŒ‰å¹´')
plt.legend()
plt.show()
```

```python
#è®¾ç½®å‚æ•°èŒƒå›´
ps = range(0,3)
qs = range(0,3)
parameters = product(ps,qs)
parameters_list = list(parameters)

#å¯»æ‰¾æœ€ä¼˜ARMAæ¨¡åž‹å‚æ•°ï¼Œå³best_aicæœ€å°
results = []
best_aic = float('inf')  #æ­£æ— ç©·
for param in parameters_list:
    try:
        model = ARMA(df_month.open,order=(param[0],param[1])).fit()
    except ValueError:
        print('å‚æ•°é”™è¯¯ï¼š',param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param,model.aic])



#è¾“å‡ºæœ€ä¼˜æ¨¡åž‹
result_table = pd.DataFrame(results)
result_table.columns = ['parameters','aic']
print('æœ€ä¼˜æ¨¡åž‹ï¼š',best_model.summary())
```

```python

df_month2 = df_month[['open']]
data_list = [datetime(2018,11,30),datetime(2018,12,31),datetime(2019,1,31),datetime(2019,2,28),datetime(2019,3,31),datetime(2019,4,30),datetime(2019,5,31),datetime(2019,6,30),datetime(2019,7,29)]
future = pd.DataFrame(index = data_list,columns = df_month.columns)
df_month2 = pd.concat([df_month2,future])
df_month2['forecast'] = best_model.predict(start=0,end = 100)


#æ¯”ç‰¹å¸é¢„æµ‹ç»“æžœæ˜¾ç¤º
plt.figure(figsize=[20,10])
df_month2.open.plot(label='å®žé™…é‡‘é¢')
df_month2.forecast.plot(color = 'r',ls = '--',label= 'é¢„æµ‹é‡‘é¢')
plt.legend()
plt.title('æ¯”ç‰¹å¸å¼€ç›˜ä»·ï¼ˆæœˆï¼‰')
plt.xlabel('æ—¶é—´')
plt.ylabel('ç¾Žé‡‘')
plt.show()
```
