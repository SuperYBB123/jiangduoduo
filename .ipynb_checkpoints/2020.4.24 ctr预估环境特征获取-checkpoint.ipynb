```python
import pandas as pd
import numpy as np
```

```python
origin_data=pd.read_csv(r'F:\work\4.20 ctr预估\origin_data.csv')
area=pd.read_csv(r'F:\work\4.20 ctr预估\2019年省市县区数据Excel(2019年10月更新).csv',encoding='gbk')
```

```python
data1=pd.merge(origin_data,area,how='inner',on='areacode')
rate=pd.read_csv(r'F:\work\4.20 ctr预估\click+action.csv')
rate=rate.drop(columns=['expose','click','action'])
df=pd.merge(data1,rate,how='left',on='itemid')
df
```

```python
df['days']=(pd.to_datetime(df['begintime'])-pd.to_datetime(df['d_dateline'])).dt.days
df['hour']=pd.to_datetime(df['begintime']).dt.hour
df['device_type']=df['channel'].map(lambda x:str(x).split('+')[0])
df['weekday']=pd.to_datetime(df['begintime']).dt.weekday_name
df['is_weekend']=df.weekday.apply(lambda x:1 if x=='Saturday' or x=='Sunday' else 0)
vocation=['2019-05-01','2019-05-02','2019-05-03','2019-05-04','2019-06-07','2019-06-08','2019-06-09',
          '2019-10-01','2019-10-02','2019-10-03','2019-10-04','2019-10-05','2019-10-06','2019-10-07',
          '2020-01-01','2020-01-24','2020-01-25','2020-01-26','2020-01-27','2020-01-28','2020-01-29',
          '2020-01-30','2020-01-31','2020-02-01','2020-02-02','2020-04-04','2020-04-05','2020-04-06']
df['action_date']=pd.to_datetime(df['begintime']).dt.date
df['is_vocation']=df.action_date.apply(lambda x:1 if x in vocation else 0)
df['a']=df['is_weekend']+df['is_vocation']
df['is_rest']=df.a.apply(lambda x:0 if x==0 else 1)
df['period_of_day']=df.hour.apply(lambda x:'上午' if 6<=x<12 else '下午' if 12<=x<18 else '凌晨' if 0<=x<6 else '晚上')
df
```

```python
df=df.drop(columns=['d_dateline','areacode','channel','p_day','area_level','hour','weekday','is_weekend',
                    'action_date','is_vocation','a','action_date'],axis=1)
df
```

```python
label_score=pd.read_csv(r'F:\work\4.20 ctr预估\用户标签得分.csv')
df2=pd.merge(df,label_score,how='left',on='deviceid')
df2
```

```python
first_city=['北京市','上海市','广州市','深圳市','成都市','杭州市','重庆市','武汉市','西安市','苏州市','天津市','南京市',
           '长沙市','郑州市','东莞市','青岛市','沈阳市','宁波市','昆明市']
second_city=['无锡市','佛山市','合肥市','大连市','福州市','厦门市','哈尔滨市','济南市','温州市','南宁市','长春市','泉州市',
            '石家庄市','贵阳市','南昌市','金华市','常州市','南通市','嘉兴市','太原市','徐州市','惠州市','珠海市','台州市',
            '中山市','烟台市','兰州市','绍兴市','海口市','扬州市']
df2['city_level']=df2.city.apply(lambda x:1 if x in first_city else 2 if x in second_city else 3)
df2
```

```python
user_city=pd.read_csv(r'F:\work\4.20 ctr预估\user_city.csv')
df3=pd.merge(df2,user_city,how='left',on='deviceid')
df4=pd.merge(df3,area,how='left',on='areacode')
df4=df4.drop(columns=['areacode','town_y','area_level'])
df4['city_level']=df4.city_y.apply(lambda x:1 if x in first_city else 2 if x in second_city else 3)
df4.rename(columns={'province_y':'user_province','city_y':'user_city','city_level':'user_city_level'},inplace=True)
df4
```

```python
register_time=pd.read_csv(r'F:\work\4.20 ctr预估\register_time.csv')
df5=pd.merge(df4,register_time,how='left',on='deviceid')
df5['a']=df5['register_days'].fillna(0)
df5['registerdays']=df5.a.apply(lambda x:int(x) if int(x)>=0 else 0)
df5=df5.drop(columns=['register_days','a'])
df5
```

```python
user_sex=pd.read_csv(r'F:\work\4.20 ctr预估\sex.csv')
df6=pd.merge(df5,user_sex,how='left',on='deviceid')
df6
```

```python
user_character=pd.read_csv(r'F:\work\4.20 ctr预估\用户特征_7.csv')
df7=pd.merge(df6,user_character,how='left',on='deviceid')
df7
```

```python
user_character_15=pd.read_csv(r'F:\work\4.20 ctr预估\用户特征_15.csv')
user_character_15=user_character_15.fillna(0)
df8=pd.merge(df7,user_character_15,how='left',on='deviceid')
df8
```
