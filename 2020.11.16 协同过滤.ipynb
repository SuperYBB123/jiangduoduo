```python
file=open(r'C:\Users\JDD\rating.csv','r')
data={}
for line in file.readlines()[1:]:
    line=line.strip().split(',')
    if not line[0] in data.keys():
        data[line[0]] = {line[1]:line[2]}
    else:
        data[line[0]][line[1]] = line[2]
data
```

```python
from math import *
#计算欧氏距离
def Euclidean(user1,user2):
    user1_data=data[user1]
    user2_data=data[user2]
    distance = 0
    for key in user1_data.keys():
        if key in user2_data.keys():
            distance += pow(float(user1_data[key])-float(user2_data[key]),2)
    return 1/(1+sqrt(distance))
```

```python
def jaccard(user1,user2):
    user1_data=data[user1]
    user2_data=data[user2]
    common=[]
    for essay in user1_data.keys():
        if essay in user2_data.keys() and int(user2_data[essay])>0 and int(user1_data[essay])>0:
            common.append(essay)
    if len(common) == 0:
        return 0
    n=len(common)
    sum1=len(user1_data.keys())
    sum2=len(user2_data.keys())
    jaccard=n/sqrt(sum1*sum2)
    return jaccard
```

```python
def pearson_sim(user1,user2):
        user1_data = data[user1]
        user2_data = data[user2]
        distance = 0
        common = {}
        for key in user1_data.keys():
            if key in user2_data.keys():
                common[key] = 1
        if len(common) == 0:
            return 0
        n = len(common)
        sum1 = sum([float(user1_data[essay]) for essay in common])
        sum2 = sum([float(user2_data[essay]) for essay in common])
        sum1Sq = sum([pow(float(user1_data[essay]),2) for essay in common])
        sum2Sq = sum([pow(float(user2_data[essay]),2) for essay in common])
        PSum = sum([float(user1_data[it])*float(user2_data[it]) for it in common])
        num = PSum - (sum1*sum2/n)
        den = sqrt((sum1Sq-pow(sum1,2)/n)*(sum2Sq-pow(sum2,2)/n))
        if den == 0:
            return 0
        r = num/den
        return r
```

```python
#计算用户相似度，找到相似的用户
def top_simliar(deviceid):
    res = []
    for userid in data.keys():
        if not userid == deviceid:
            simliar = jaccard(deviceid,userid)
            res.append((userid,simliar))
    res.sort(key=lambda val:val[1],reverse=True)
    return res[:20]
```

```python
#根据用户推荐文章
def recommend(user):
    top_sim_users=[]
    for res in top_simliar(user):
        top_sim_users.append(res[0])
    for top_sim_user in top_sim_users:
        items = data[top_sim_user]
        recommendations = []
        for item in items.keys():
            if item not in data[user].keys() and item not in recommendations and int(items[item])>0:
                recommendations.append((item,items[item]))
    recommendations.sort(key=lambda val:val[1],reverse=True)
    return recommendations[:10]

Recommendations = recommend('92bb9accb1c94902a0234ecb8cf1a44e')
print(Recommendations)
```

```python
for userid in data.keys():
    recommendations=recommend(userid)
    print(userid+':'+str(recommendations))
```
