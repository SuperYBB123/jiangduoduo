```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv(r'C:\Users\JDD\Downloads\query-impala-346925.csv')
df
```

```python
output=pd.DataFrame(columns=['auther','1','2','3','pct1','pct2','pct3'])
```

```python
output
```

```python
auther_list=[]
for auther in df['n_autherid'].unique():
    auther_list.append(str(auther))
len(auther_list)
```

```python
for auther in auther_list:
    auther=int(auther)
    a=len(df.loc[(df['n_autherid']==auther) & (df['behavior_type']==1)])
    b=len(df.loc[(df['n_autherid']==auther) & (df['behavior_type']==2)])
    c=len(df.loc[(df['n_autherid']==auther) & (df['behavior_type']==3)])
    d=len(df.loc[df['n_autherid']==auther])
    pct1=a/d
    pct2=b/d
    pct3=c/d
    new=pd.DataFrame({'auther':auther,'1':a,'2':b,'3':c,'pct1':pct1,'pct2':pct2,'pct3':pct3},index=[1])
    output=output.append(new,ignore_index=True)
```

```python
output
```
