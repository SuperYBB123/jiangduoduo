```python
import warnings
warnings.filterwarnings('ignore')#忽略一些警告
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
columns = ['date', 'price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 
           'floors', 'grade', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 
           'lat', 'long']

kc_train = pd.read_csv('./kc_train.csv', names=columns)
```

```python
print("Skewness: %f" %kc_train['price'].skew())      #偏度
print("Kurtosis: %f" %kc_train['price'].kurt())      #峰度
Skewness: 3.898737
Kurtosis: 29.356202
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['font.sans-serif']=['SimHei'] 
plt.rcParams['axes.unicode_minus']=False

corrmat = kc_train.corr()         
plt.subplots(figsize=(12,9))
sns.heatmap(corrmat, vmax=0.9, square=True,center=2.0)
```

```python
k = 10 
plt.figure(figsize=(12,9))
cols = corrmat.nlargest(k, 'price')['price'].index
cm = np.corrcoef(kc_train[cols].values.T)
sns.set(font_scale=1.25)
hm = sns.heatmap(cm, cbar=True, annot=True, square=True, fmt='.2f', annot_kws={'size': 10}, yticklabels=cols.values, xticklabels=cols.values)
plt.show()
```
