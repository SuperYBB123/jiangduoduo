```python
import pandas as pd
import math
import random
import numpy as np
```

```python
tmp=pd.read_csv(r'F:\work\10.15 simrank线上线下联调\8-14.csv')
scan_dict={}
for i in tmp.index:
    deviceid=tmp['deviceid'][i]
    scan_dict.setdefault(deviceid,{})
    j=tmp['list'][i]
    k=j.strip().split(',')
    goal=k[-5:]
    goal.reverse()
    for item in goal:
        score=5-int(goal.index(item))
        if item not in scan_dict[deviceid].keys():
            scan_dict[deviceid][item]=score
        else:
            old_score=scan_dict[deviceid][item]
            scan_dict[deviceid][item]=max(score,old_score)
```

```python
test={}
tmp=pd.read_csv(r'F:\work\10.15 simrank线上线下联调\15expose.csv')
for i in tmp.index:
    deviceid=tmp['deviceid'][i]
    test.setdefault(deviceid,{})
    j=tmp['list'][i]
    k=j.strip().split(',')
    test[deviceid]['expose']=k
tmp=pd.read_csv(r'F:\work\10.15 simrank线上线下联调\15click.csv')
for i in tmp.index:
    deviceid=tmp['deviceid'][i]
    if deviceid in test.keys():
        j=tmp['list'][i]
        k=j.strip().split(',')
        test[deviceid]['click']=k
```

```python
sim_dict={}
tmp=pd.read_csv(r'F:\work\10.15 simrank线上线下联调\simrank14.csv')
for i in tmp.index:
    item_a=tmp['item_a'][i]
    item_b=tmp['item_b'][i]
    score=tmp['score'][i]
    sim_dict.setdefault(item_a,{})
    sim_dict[item_a][item_b]=score
```

```python
Sim_dict={}
for i in sim_dict.keys():
    max_score=np.max(list(sim_dict[i].values()))
    min_score=np.min(list(sim_dict[i].values()))
    Sim_dict.setdefault(str(i),{})
    for j in sim_dict[i].keys():
        score=sim_dict[i][j]
        new_score=(score-min_score)/(max_score-min_score)
        Sim_dict[str(i)][str(j)]=new_score
```

```python
device_list=[x for x in scan_dict.keys() if x in test.keys()]
random.seed(23)
sample_device=random.sample(device_list,20000)
ACC=[]
RECALL=[]
expose_ACC=[]
for device in sample_device:
    scan=scan_dict[device]
    rec_dict={}
    for item in scan.keys():
        scan_score=scan[item]
        if item in Sim_dict.keys():
            for item_b in Sim_dict[item].keys():
                sim_score=Sim_dict[item][item_b]
                sum_score=sim_score*scan_score
                if item_b in rec_dict.keys():
                    old_score=rec_dict[item_b]
                    rec_dict[item_b]=max(old_score,sum_score)
                else:
                    rec_dict[item_b]=sum_score
    rec_sorted=sorted(rec_dict.items(),key=lambda x:x[1],reverse=True)[:50]
    rec=[x[0] for x in rec_sorted]
    exposed=[x for x in test[device]['expose'] if x in rec]
    if len(exposed)>0:
        if len(test[deviceid])==2:
            true=[x for x in rec if x in test[deviceid]['click']]
            po=[x for x in test[deviceid]['click'] if x in rec]
            acc=len(true)/len(rec)
            recall=len(po)/len(test[deviceid]['click'])
            expose_acc=len(true)/len(exposed)
#        else:
#            acc=0
#            recall=0
#            expose_acc=0
            ACC.append(acc)
            RECALL.append(recall)
            expose_ACC.append(expose_acc)
print(np.mean(ACC),np.mean(RECALL),np.mean(expose_ACC),2*np.mean(RECALL)*np.mean(expose_ACC)/(np.mean(RECALL)+np.mean(expose_ACC)),len(ACC))
```
