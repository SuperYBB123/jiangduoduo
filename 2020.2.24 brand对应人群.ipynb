```python
brand_list=list(df['brand_id'].unique())
len(brand_list)
```

```python
output=pd.DataFrame(columns=['brand','car_cnt','device_cnt','essay_cnt'])
for brand in brand_list:
    device_cnt=len(df.loc[df['brand_id']==brand].deviceid.unique())
    essay_cnt=len(df.loc[df['brand_id']==brand].essay_id.unique())
    car_cnt=len(df.loc[df['brand_id']==brand].car_id.unique())
    new=pd.DataFrame({'brand':brand,'car_cnt':car_cnt,'device_cnt':device_cnt,'essay_cnt':essay_cnt},index=[1])
    output=output.append(new,ignore_index=True)
output
```
