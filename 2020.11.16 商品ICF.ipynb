```python
import pandas as pd
import numpy as np
import random
import math
```

```python
df=pd.read_csv(r'C:\Users\JDD\Downloads\download (19).csv',encoding='gbk')
click_data={}
purchase_data={}
goods_info={}
for i in df.index:
    deviceid=df['deviceid'][i]
    goods_id=str(int(df['goods_id'][i]))
    goods_name=df['goods_name'][i]
    eventid=df['eventid'][i]
    first_category=df['first_category_name'][i]
    second_category=df['second_category_name'][i]
    price=df['sell_price'][i]
    if eventid=='P_SC0432':
        click_data.setdefault(deviceid,{})
        click_data[deviceid][goods_id]=1
    else:
        purchase_data.setdefault(deviceid,{})
        purchase_data[deviceid][goods_id]=1
    goods_info.setdefault(goods_id,{})
    goods_info[goods_id]['goods_name']=goods_name
    goods_info[goods_id]['first_category']=first_category
    goods_info[goods_id]['second_category']=second_category
    goods_info[goods_id]['price']=price
```

```python
train=purchase_data
C=dict()  # item-iten的共现矩阵
N=dict()  # item被多少个不同用户点击
for user,items in train.items():
    for i in items.keys():
        N.setdefault(i,0)
        N[i]+=1
        C.setdefault(i,{})
        for j in items.keys():
            if i==j: continue
            C[i].setdefault(j,0)
            C[i][j]+=1
similarity=dict()  # 相似度矩阵
for i,related_items in C.items():
    similarity.setdefault(i,{})
    for j,cij in related_items.items():
        if cij>=10:
            similarity[i][j]=cij/(math.sqrt(N[i]*N[j]))
```

```python
output=[]
for e1,x in similarity.items():
    x_sorted=sorted(x.items(), key=lambda x:x[1], reverse=True)[:50]
    for y in x_sorted:
        e2=y[0]
        if e1 in goods_info.keys() and e2 in goods_info.keys():
            output.append((e1,goods_info[e1]['goods_name'],
                           goods_info[e1]['first_category'],goods_info[e1]['second_category'],goods_info[e1]['price'],
                           e2,goods_info[e2]['goods_name'],
                           goods_info[e2]['first_category'],goods_info[e2]['second_category'],goods_info[e2]['price'],
                           x[e2]))
```

```python
columns=['goods_1','name_1','1st_cate_1','2nd_cate_1','price_1',
        'goods_2','name_2','1st_cate_2','2nd_cate_2','price_2','sim']
pd.DataFrame(output,columns=columns).to_csv(r'F:\work\5.18 word2vec训练文章id\车辆id\1118商品icf购买1月10.csv',index=False)
```
