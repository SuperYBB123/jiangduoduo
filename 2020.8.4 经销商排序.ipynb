```python
df=pd.read_csv(r'F:\work\8.3 经销商排序\经销商位置点击曝光.csv')
shop_position_expose=dict()
shop_position_click=dict()
for i in df.index:
    shop=df['shop_id'][i]
    position=df['rec_position'][i]
    clicks=int(df['click_pv'][i])
    exposes=int(df['exposure_pv'][i])
    shop_position_expose.setdefault(shop,{})
    shop_position_expose[shop][position]=exposes
    shop_position_click.setdefault(shop,0)
    shop_position_click[shop]+=clicks
shop_position_expose
```

```python
def mAP_expose(user,n):
    user_province=user_info[user]['province']
    user_city=user_info[user]['city']
    click_list=list(test_click[user].keys())
    expose_list=list(test_expose[user].keys())
    city_shop_list=list(sorted_city_shop[user_province][user_city].keys())[:n]
    rec=[x for x in city_shop_list if x in expose_list]
    precision_list=[]
    recall_list=[0]
    for i in range(0,len(rec)):
        reco=rec[:i+1]
        precision=len([x for x in reco if x in click_list])/len(reco)
        recall=len([x for x in click_list if x in reco])/len(click_list)
        if recall>np.max(recall_list):
            recall_list.append(recall)
            precision_list.append(precision)
    if len(precision_list)==0:
        precision_list.append(0)
    return np.mean(precision_list),rec,precision_list,len(rec)
```
