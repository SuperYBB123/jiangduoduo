```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
lamda=3
t=0
I=0
i=[]
S_II=[]
while t<=10:
    R=np.random.uniform(0, 1)
    t=t+(-1/lamda)*np.log(R)
    I=I+1
    S_I=t
    S_II.append(S_I)
    i.append(I)
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号
plt.figure(figsize = (6, 4)) #设置图像大小
plt.xlabel(u"时间t")#横轴
plt.ylabel(u"S(I)")#纵轴
plt.plot(S_II, i,color = 'black', linewidth = 2)
plt.show()
```

```python
import numpy as np
import pandas as pd
np.random.seed(1234)
x=np.random.random(50)#生成0，1之间的50个随机数
p0=[1/3,1/3,1/3]
P=np.matrix([[0.2,0.3,0.5],[0.7,0.2,0.1],[0.5,0.1,0.4]])
if x[0]<=1/3:
    Xn=1
elif 1/3<x[0]<=2/3:
    Xn=2
else:
    Xn=3
X_n=[]
for i in range(1,50):
    if Xn==1:
        if x[i]<=P[0,0]:
            Xn=1
        elif P[0,0]<x[i]<=P[0,0]+P[0,1]:
            Xn=2
        else:
            Xn=3
    elif Xn==2:
        if x[i]<=P[1,0]:
            Xn=1
        elif P[1,0]<x[i]<=P[1,0]+P[1,1]:
            Xn=2
        else:
            Xn=3
    elif Xn==3:
        if x[i]<=P[2,0]:
            Xn=1
        elif P[2,0]<x[i]<=P[2,0]+P[2,1]:
            Xn=2
        else:
            Xn=3
    i+=1
    X_n.append(Xn)
X_n
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(1234)
n=200
deltat=1/n
t=0
t_j=[]
while t<=1:
    print(t)
    t+=deltat
    t_j.append(t)
z=np.random.normal(loc=0.0, scale=1.0, size=200)
deltaw=z*np.sqrt(deltat)
z_j=z.tolist()
deltaw=pd.DataFrame(deltaw)
deltaw_jj=[]
for j in range(n):
    deltaw_j=deltaw.iloc[j,0]
    deltaw_jj.append(deltaw_j)
deltaw_jj=pd.DataFrame(deltaw_jj)
w_j=0
w_jj=[]
for i in range(n):
    w_j=w_j+deltaw_jj.iloc[i,0]
    w_jj.append(w_j)
plt.rcParams['font.sans-serif'] = ['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False #用来正常显示负号
plt.figure(figsize = (8, 4)) #设置图像大小
plt.title('标准布朗过程') #标题
plt.xlabel(u"时间")#横轴
plt.ylabel(u"w")#纵轴
plt.plot(t_j,w_jj)
plt.show()
```

```python
import pandas as pd
import numpy as np
from numpy.linalg import cholesky
W=np.matrix([[4,0.8,0.16],[0.8,4,0.8],[0.16,0.8,4]])
a=np.matrix([[1],[1.5],[2]])
C=cholesky(W)
np.random.seed(666)
u1=np.random.normal(loc=0.0, scale=1.0, size=100)
np.random.seed(1234)
u2=np.random.normal(loc=0.0, scale=1.0, size=100)
np.random.seed(4321)
u3=np.random.normal(loc=0.0, scale=1.0, size=100)
x1=a[0,0]+C[0,0]*u1
x2=a[1,0]+C[1,0]*u1+C[1,1]*u2
x3=a[2,0]+C[2,0]*u1+C[2,1]*u2+C[2,2]*u3
x1,x2,x3=x1.tolist(),x2.tolist(),x3.tolist()
X_I=[]
for i in range(100):
    X=[x1[i],x2[i],x3[i]]
    X_I.append(X)
X_I
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

N =5000
theta[0] = 0.2
U = np.random.rand(N,1)
for i in range(0,N-1):
    #y =np.random.normal(loc = theta[i],scale = 0.01, size=1)
    #y =np.random.uniform(0, 1)   
    y=np.random.normal(0.4,scale = 0.01, size=1)
    alpha=((1/2+y/4)**125*(1-y)**38*(y/4)**34)/((1/2+theta[i]/4)**125*(1-theta[i])**38*(theta[i]/4)**34)  
    if U[i]<alpha:
        theta[i+1]= y
    else:
        theta[i+1]=theta[i]

sns.distplot(theta, kde = True,color = 'blue')
plt.title('N(0.4,0.01)')
plt.xlabel('theta')
plt.ylabel('Frequency')
plt.show()
print(np.mean(theta[4000:]))
```
