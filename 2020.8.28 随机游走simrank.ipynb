```python
import pandas as pd
import numpy as np
import random
import datetime
```

```python
iteration=5
nodes = []  # 所有的节点存入数组
nodesnum = 0  # 所有节点的数目
nodes_index = {}  # <节点名，节点在nodes数组中的编号>
link_in={}
trans_matrix=np.matrix(0)
damp = 0.8  # 阻尼系数
simmatrix=np.matrix(0)
df=pd.read_csv(r'F:\work\8.19 simrank\title.csv')
df=df.loc[df['c_title'].notna()]
title_dict=dict()
for i in df.index:
    itemid=df['itemid'][i]
    title=df['c_title'][i]
    title_dict[str(itemid)]=title
```

```python
def initParam():
    '''
    构建nodes、nodes_index、trans_matrix和第0代的sim_matrix.
    输入文件行格式要求：node\toutneighbor\toutneighbor\t...或 node\tinneighbor\tinneighbor\t...
    '''
    global nodes
    global nodes_index
    global sim_matrix
    global damp
    global nodesnum
    global link_in
    global trans_matrix
 
    link_in = {}
    df=pd.read_csv(r'F:\work\8.19 simrank\7天user.csv')
    df=df[:10000]
    for i in df.index:
        node = df['deviceid'][i]
        item_list=df['list'][i].strip().split(',')
        nodeid = -1
        if node in nodes_index:
            nodeid = nodes_index[node]
        else:
            nodeid = len(nodes)
            nodes_index[node] = nodeid
            nodes.append(node)
        link_in.setdefault(node,[])
        for item in item_list:
            if item not in link_in[node]:
                link_in[node].append(item)
                
                
    for i in df.index:
        node = df['deviceid'][i]
        nodeid = nodes_index[node]
        item_list=df['list'][i].strip().split(',')
        for ele in item_list:
            outneighbor = ele
            outneighborid = -1
            if outneighbor in nodes_index:
                outneighborid = nodes_index[outneighbor]
            else:
                outneighborid = len(nodes)
                nodes_index[outneighbor] = outneighborid
                nodes.append(outneighbor)
            link_in.setdefault(outneighbor,[])
            if node not in link_in[outneighbor]:
                link_in[outneighbor].append(node)
                
    nodesnum = len(nodes)
    sim_matrix = np.identity(nodesnum)
    trans_matrix = np.zeros((nodesnum, nodesnum))
    for node, inneighbors in link_in.items():
        nodeid=nodes_index[node]
        num = len(inneighbors)
        prob = 1.0 / num
        for neighbor in inneighbors:
            neighbourid=nodes_index[neighbor]
            trans_matrix[nodeid,neighbourid] = prob
```

```python
def RandomWalk():
    
    global nodes
    global nodes_index
    global sim_matrix
    global damp
    global nodesnum
    global link_in
    global iteration
    global trans_matrix
    
    routes={}
    for p1 in link_in.keys():
        tmp=[p1]
        routes.setdefault(p1,[])
        p1_index=nodes_index[p1]
        for neighbour in link_in[p1]:
            a=tmp
            a.append(neighbour)
            routes[p1].append(a)
```

```python
for p1 in link_in.keys():
    p1id=nodes_index[p1]
    p1neighbors=link_in[p1]
    for p2 in link_in.keys():
        p2id=nodes_index[p2]
        if p1id<p2id:
            p2neighbors=link_in[p2]
            common=[x for x in p1neighbors if x in p2neighbors]
            new_sim=len(common)*len(common)/(len(p1neighbors)*len(p2neighbors))
            sim_matrix[p1id,p2id]+=new_sim
            sim_matrix[p2id,p1id]+=new_sim
```

```python
for k in range(4):
    tmp_matrix=trans_matrix
    for p1 in link_in.keys():
        p1id=nodes_index[p1]
        for p2 in link_in.keys():
            p2id=nodes_index[p2]
            for p3 in link_in[p2]:
                p3id=nodes_index[p3]
                p1_p2_trans=trans_matrix[p1id,p2id]
                new_trans=p1_p2_trans*(1/len(link_in[p2]))
                trans_matrix[p1id,p3id]+=new_trans
```
