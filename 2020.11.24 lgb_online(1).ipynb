```python
import numpy as np
import pandas as pd
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score
from sklearn.metrics import recall_score  
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#连续特征列
continuous_col = ['n_activate_utilnow','d_st_ctime','d_st_titlelen',
                  'd_dy_histpv','d_dy_recvalidshow','d_dy_recread','d_dy_recclickrate',
                  'd_dy_finishrate','n_browse_price','u_dy_rt_poscnts','u_dy_rt_videoreadratio','d_st_playtime','d_st_contentlen']#,'u_dy_rt_readtime',,'exposure_times',
                 #'click_times','favorite_times','like_times','reply_times','display_ctr_no','ctr_no','two_week_active_days',
                 #'two_week_page_avg_view','view_essay_two_week_avg_count','view_moment_two_week_avg_count',
                 # 'view_car_two_week_avg_count','level_1','level_2','level_3'
#离散特征列
discrete_col = ['plateform','province','city_level','is_rest_day','is_daytime','d_st_contenttype',
                'd_st_covershowtype','d_st_life','d_dy_isexpired','position','context']
#string型特征
string_col = ['d_st_tag','u_dy_rt_notclickedtag','u_dy_rt_readtag','u_dy_rt_poscate','s_long_labels','s_short_labels']
```

```python
file_loc_x = './1107-1109线上+线下特征.csv'
data = pd.read_csv(file_loc_x)
#删除实时特征
data = data[continuous_col+discrete_col+string_col+['deviceid','action_type']]
data.dropna(axis=0,subset=['u_dy_rt_notclickedtag','u_dy_rt_readtag','u_dy_rt_poscnts',
                           'u_dy_rt_poscate','u_dy_rt_videoreadratio'],how='all',inplace=True)

file_loc_y = './1110线上+线下特征.csv'
data_y = pd.read_csv(file_loc_y)
data_y = data_y[continuous_col+discrete_col+string_col+['deviceid','action_type']]
#删除实时特征
data_y.dropna(axis=0,subset=['u_dy_rt_notclickedtag','u_dy_rt_readtag','u_dy_rt_poscnts',
                           'u_dy_rt_poscate','u_dy_rt_videoreadratio'],how='all',inplace=True)
indice = data.shape[0]
data = pd.concat([data,data_y],axis=0)
data.reset_index(drop=True, inplace=True)
```

```python
def string_fillna(data):
    #空值处理
    data['d_st_tag']=data['d_st_tag'].fillna('null')
    #空值处理
    data['u_dy_rt_notclickedtag']=data['u_dy_rt_notclickedtag'].fillna('null')
    #空值处理
    data['u_dy_rt_readtag']=data['u_dy_rt_readtag'].fillna('null')
    #空值处理
    data['u_dy_rt_poscate']=data['u_dy_rt_poscate'].fillna('6')
    #data['u_dy_rt_poscate'] = [str (item) for item in data['u_dy_rt_poscate']]
    #空值处理
    data['s_long_labels']=data['s_long_labels'].fillna('null')
    #空值处理
    data['s_short_labels']=data['s_short_labels'].fillna('null')
    
string_fillna(data)
#string_fillna(data_y)
```

```python
#划分测试集
#x_train_all, x_predict, y_train_all, y_predict = train_test_split(X, y, test_size=0.10, random_state=100)
x_train_all, x_predict, y_train_all, y_predict = X[:indice],X[indice:],y[:indice],y[indice:]
x_train, x_test, y_train, y_test = train_test_split(x_train_all, y_train_all, test_size=0.2, random_state=100)

#dataset包装
train_data = lgb.Dataset(data=x_train, label=y_train, feature_name = feature_names, categorical_feature = cate_list)
test_data = lgb.Dataset(data=x_test, label=y_test, feature_name = feature_names, categorical_feature = cate_list)
```

```python
#参数设置
params = {
    'boosting_type': 'gbdt',  # 设置提升类型
    'objective': 'binary',  # 目标函数
    'metric': {'binary_logloss', 'auc'},  # 评估函数
    'num_trees': 100,
    'max_depth': 8,
    'num_leaves': 128,  # 叶子节点数
    'learning_rate': 0.1,  # 学习速率
    'feature_fraction': 0.8,  # 建树的特征选择比例
    'bagging_fraction': 0.8,  # 建树的样本采样比例
    'bagging_freq': 5,  # 每 k 次迭代执行bagging
    'min_data_in_leaf': 500,
    'max_bin': 128,
    'verbose': 1,  # <0 显示致命的, =0 显示错误 (警告), >0 显示信息
    'is_unbalance' : True,
    'min_data_in_bin': 50
}
```

```python
num_round = 10
lgb_1 = lgb.train(params, train_data, num_round, valid_sets=[test_data])
```

```python
#预测与模型评估
y_pred = lgb_1.predict(x_predict)
y_pred_binary = np.where(y_pred > 0.5, 1, 0) 
acc = accuracy_score(y_predict, y_pred_binary)
recall = recall_score(y_predict, y_pred_binary) 
auc = roc_auc_score(y_predict, y_pred)
print('auc:', auc, 'acc:', acc, 'recall', recall)
```

```python
#特征重要性可视化
import matplotlib.pylab as plt
import seaborn as sns
plt.rcParams['font.sans-serif'] = ['FangSong'] #中文乱码

def lgb_importance():
    
    
    feature_names_pd = pd.DataFrame({'column': feature_names,
                                     'importance': lgb_1.feature_importance(importance_type = 'gain'),
                                     }).sort_values(by='importance')
    print(feature_names_pd)
    cols = feature_names_pd[["column", "importance"]].groupby("column").mean().sort_values(by="importance",ascending=False)[:50].index
                                         
    best_features = feature_names_pd.loc[feature_names_pd.column.isin(cols)]
    
    plt.figure(figsize=(10, 15))
    sns.barplot(x="importance", y="column", data=best_features.sort_values(by="importance", ascending=False))
    plt.title('LightGBM Features')
    plt.tight_layout()

lgb_importance()
```
