```python
import pandas as pd
from gensim.models import word2vec
import numpy as np
```

```python
tmp=pd.read_csv(r'C:\Users\JDD\Downloads\11.16车辆word2vec.csv')
fl = open(r'F:\work\5.18 word2vec训练文章id\车辆id\data.txt', 'w', encoding='utf8')
for i in tmp.index:
    j=tmp['list'][i]
    if pd.isnull(j)==False:
        k=j.strip().split(',')
        for wd in k:
            fl.write(wd)
            fl.write(' ')
        fl.write('\n')
fl.close()
```

```python
tmp=pd.read_csv(r'C:\Users\JDD\Downloads\query-impala-497907.csv')
effective_cars={}
for i in tmp.index:
    car=str(tmp['goods_id'][i])
    brand_name=tmp['brand_name'][i]
    goods_car_name=tmp['goods_name'][i]
    effective_cars.setdefault(car,{})
    effective_cars[car]['brand_name']=brand_name
    effective_cars[car]['goods_name']=goods_car_name
```

```python
sentences=word2vec.Text8Corpus(r'F:\work\5.18 word2vec训练文章id\车辆id\data.txt')
model=word2vec.Word2Vec(sentences,size=30,min_count=40,window=6,compute_loss=True,
                  workers=6,sg=0,sample=1e-4,hs=0,negative=40,iter= 1,sorted_vocab=1)
```

```python
cars=list(model.wv.vocab.keys())
car_sim_degree=dict()
for car in cars:
    car_sim_degree.setdefault(car,{})
    for i,x in enumerate(model.wv.most_similar(car,topn=50)):
        sim=x[0]
        degree=x[1]
        car_sim_degree[car][sim]=degree
```

```python
output=[]
for e1,x in car_sim_degree.items():
    for e2 in x.keys():
        if e1 in effective_cars.keys() and e2 in effective_cars.keys():
            output.append((e1,effective_cars[e1]['brand_name'],effective_cars[e1]['goods_name'],
                           e2,effective_cars[e2]['brand_name'],effective_cars[e2]['goods_name'],x[e2]))
print('相似度获取成功!')
print('输出长度：',len(output))
```

```python
pd.DataFrame(output).to_csv(r'F:\work\5.18 word2vec训练文章id\车辆id\1116carsim2.csv',index=False)
```
